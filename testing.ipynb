```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'd:\\IPL__DATASET\\archive\\matches.csv'
df = pd.read_csv(file_path)

# Filter for final matches
finals_df = df[df['match_type'] == 'Final']

# Get the winners of the finals
title_winners = finals_df['winner']

# Count the number of titles for each team
title_counts = title_winners.value_counts()

# Create a color palette
colors = sns.color_palette('viridis', len(title_counts))

# Create the bar plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=title_counts.index, y=title_counts.values, palette=colors)

# Add titles and labels
plt.title('IPL Title Winners (2008-2024)', fontsize=18)
plt.xlabel('Teams', fontsize=14)
plt.ylabel('Number of Titles', fontsize=14)
plt.xticks(rotation=45, ha='right')

# Add data labels on top of the bars
for i, v in enumerate(title_counts.values):
    ax.text(i, v + 0.1, str(v), color='black', ha='center', fontweight='bold')

# Add a background grid
ax.yaxis.grid(True, linestyle='--', linewidth=0.5)
ax.set_axisbelow(True)

# Add a border around the plot
for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(1)

# Save the plot
output_path = 'd:\\IPL__DATASET\\ipl_title_winners.png'
plt.savefig(output_path, bbox_inches='tight')

print(f"Visualization saved to: {output_path}")

plt.show()
```

```python
# 1. Team with the most wins
most_wins = df['winner'].value_counts().idxmax()
print(f"\n1. Team with most wins: {most_wins}")

# 2. Team with the highest winning percentage
total_matches = pd.concat([df['team1'], df['team2']]).value_counts()
total_wins = df['winner'].value_counts()
win_percentage = (total_wins / total_matches) * 100
highest_win_percentage_team = win_percentage.idxmax()
highest_win_percentage_value = win_percentage.max()
print(f"\n2. Team with highest win percentage: {highest_win_percentage_team} ({highest_win_percentage_value:.2f}%)")

# 3. Biggest win margin by runs
biggest_run_margin = df[df['result'] == 'runs']['result_margin'].max()
biggest_run_margin_winner = df[df['result_margin'] == biggest_run_margin]['winner'].iloc[0]
print(f"\n3. Biggest win margin by runs: {biggest_run_margin} runs by {biggest_run_margin_winner}")

# 4. Biggest win margin by wickets
biggest_wicket_margin = df[df['result'] == 'wickets']['result_margin'].max()
biggest_wicket_margin_winner = df[df['result_margin'] == biggest_wicket_margin]['winner'].iloc[0]
print(f"\n4. Biggest win margin by wickets: {biggest_wicket_margin} wickets by {biggest_wicket_margin_winner}")

# 5. Player with most "Player of the Match" awards
most_pom = df['player_of_match'].value_counts().idxmax()
print(f"\n5. Player with most 'Player of the Match' awards: {most_pom}")

# 6. Team that won the most tosses
most_toss_wins = df['toss_winner'].value_counts().idxmax()
print(f"\n6. Team that won the most tosses: {most_toss_wins}")

# 7. Analysis of toss decisions (bat vs. field)
toss_decision_percentage = df['toss_decision'].value_counts(normalize=True) * 100
print("\n7. Toss decision analysis:")
print(toss_decision_percentage)

# 8. Impact of winning the toss on the match outcome
toss_winner_is_match_winner = (df['toss_winner'] == df['winner']).sum()
toss_winner_is_match_winner_percentage = (toss_winner_is_match_winner / len(df)) * 100
print(f"\n8. Impact of winning the toss on match outcome: Winning the toss results in a match win {toss_winner_is_match_winner_percentage:.2f}% of the time.")

# 9. City that has hosted the most matches
most_matches_city = df['city'].value_counts().idxmax()
print(f"\n9. City that has hosted the most matches: {most_matches_city}")

# 10. Most frequent umpires
umpires = pd.concat([df['umpire1'], df['umpire2']])
most_frequent_umpire = umpires.value_counts().idxmax()
print(f"\n10. Most frequent umpire: {most_frequent_umpire}")

print("\n--- Analysis Complete ---")
```

```python
import random
import pandas as pd

# Teams list
teams = [
    "Mumbai Indians", "Chennai Super Kings", "Royal Challengers Bangalore",
    "Kolkata Knight Riders", "Rajasthan Royals", "Sunrisers Hyderabad",
    "Delhi Capitals", "Punjab Kings"
]

# Venues list
venues = [
    "Wankhede Stadium", "Eden Gardens", "M Chinnaswamy Stadium",
    "Feroz Shah Kotla", "Sawai Mansingh Stadium", "Chepauk Stadium",
    "Punjab Cricket Association Stadium, Mohali", "Rajiv Gandhi Intl. Stadium"
]

# Toss options
toss_decisions = ["bat", "field"]

# Seasons list (2008 to 2023)
seasons = [str(y) for y in range(2008, 2024)]

# Generate 100 random queries
queries = []
for i in range(100):
    team1, team2 = random.sample(teams, 2)  # alag-alag teams
    venue = random.choice(venues)
    season = random.choice(seasons)
    toss_winner = random.choice([team1, team2])
    toss_decision = random.choice(toss_decisions)
    
    queries.append({
        "season": season,
        "venue": venue,
        "team1": team1,
        "team2": team2,
        "toss_winner": toss_winner,
        "toss_decision": toss_decision
    })

# Convert to DataFrame
test_queries_df = pd.DataFrame(queries)

# Show first 10 queries
print("Sample Test Queries:")
print(test_queries_df.head(10))

# Save to CSV (optional)
test_queries_df.to_csv("test_queries.csv", index=False)
print("\n 100 test queries saved to test_queries.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# 1. Season-wise query distribution
plt.figure(figsize=(12,6))
sns.countplot(data=test_queries_df, x="season", palette="viridis")
plt.title("Number of Queries per Season")
plt.xticks(rotation=45)
plt.show()

# 2. Venue frequency heatmap
plt.figure(figsize=(12,6))
sns.heatmap(test_queries_df['venue'].value_counts().to_frame().T, annot=True, cmap="coolwarm", cbar=False)
plt.title("Venue Frequency Heatmap")
plt.show()

# 3. Toss Decision Pie Chart
plt.figure(figsize=(6,6))
test_queries_df['toss_decision'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title("Toss Decision Distribution")
plt.ylabel("")
plt.show()

# 4. Team Participation Count (Bar Chart)
plt.figure(figsize=(12,6))
all_teams = pd.concat([test_queries_df['team1'], test_queries_df['team2']])
sns.countplot(x=all_teams, order=all_teams.value_counts().index, palette="cubehelix")
plt.xticks(rotation=45)
plt.title("Team Participation in Queries")
plt.show()

# 5. Interactive Sunburst Plot (Plotly)
fig = px.sunburst(test_queries_df, path=['season','venue','toss_decision'], values=None,
                  title="Query Breakdown by Season  Venue  Toss Decision",
                  color="season", color_discrete_sequence=px.colors.qualitative.Set2)
fig.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
import catboost
import joblib

# ============================
# Load Dataset
# ============================
df = pd.read_csv(r"D:\IPL__DATASET\archive\matches.csv")

# Target column (Winner)
y = df["winner"].fillna("Unknown")

# Features
X = df.drop(columns=["winner"])

# Encode categorical
for col in X.select_dtypes(include="object").columns:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str))

# Handle NaN
X.fillna(0, inplace=True)

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================
# Define Models
# ============================
log_reg = LogisticRegression(max_iter=500, random_state=42)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
svm = SVC(probability=True, random_state=42)
knn = KNeighborsClassifier(n_neighbors=7)
nb = GaussianNB()
xgb = XGBClassifier(use_label_encoder=False, eval_metric="mlogloss", random_state=42)
cat = catboost.CatBoostClassifier(verbose=0, random_state=42)

# Ensemble Model
voting_clf = VotingClassifier(
    estimators=[
        ('LogReg', log_reg),
        ('RF', rf),
        ('SVM', svm),
        ('KNN', knn),
        ('NB', nb),
        ('XGB', xgb),
        ('Cat', cat)
    ],
    voting='soft'
)

# ============================
# Train Model
# ============================
voting_clf.fit(X_scaled, y)

# Save model
joblib.dump(voting_clf, "final_model.pkl")
print(" Model trained and saved as final_model.pkl")
```

```python
# Load trained model
voting_clf = joblib.load("final_model.pkl")

# Example test query
sample = X.iloc[:5]   # pehle 5 rows test karne ke liye
sample_scaled = scaler.transform(sample)

preds = voting_clf.predict(sample_scaled)
print("Predicted Winners:", preds)
```

```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# ============================
# 1. Load Dataset + Model
# ============================
df = pd.read_csv(r"D:\IPL__DATASET\archive\matches.csv")
voting_clf = joblib.load("final_model.pkl")

# Encode categorical for consistency
df_copy = df.copy()
for col in df_copy.select_dtypes(include="object").columns:
    le = LabelEncoder()
    df_copy[col] = le.fit_transform(df_copy[col].astype(str))

# Features & scaling
X = df_copy.drop(columns=["winner"])

# Handle NaN with SimpleImputer
imputer = SimpleImputer(strategy="most_frequent")
X_imputed = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# ============================
# 2. Create 20 Test Queries
# ============================
test_queries = df.sample(20, random_state=42).drop(columns=["winner"])

# Encode queries same as dataset
for col in test_queries.select_dtypes(include="object").columns:
    test_queries[col] = LabelEncoder().fit_transform(test_queries[col].astype(str))

# Fill NaN in queries
test_queries_imputed = imputer.transform(test_queries)

# Scale queries
test_queries_scaled = scaler.transform(test_queries_imputed)

# ============================
# 3. Predictions
# ============================
predictions = voting_clf.predict(test_queries_scaled)

# ============================
# 4. Show Results
# ============================
results = pd.DataFrame({
    "Query": range(1, 21),
    "Predicted Winner": predictions
})

print(" IPL Match Predictions (20 Test Queries):")
print(results)
```
